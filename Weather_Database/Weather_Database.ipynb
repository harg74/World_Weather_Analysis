```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from citipy import citipy
from config import weather_api_key
```

```python
#Create a set of random lats y lngs combinations

lats=np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs=np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs=zip(lats,lngs)
lat_lngs
```

```python
# Add the latitudes and longitudes to a list.

coordinates = list(lat_lngs)
```

```python
cities=[]

for coordinate in coordinates:
    city=citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    if city not in cities:
        cities.append(city)
```

```python
#create URL
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial"
```

```python
city_data=[]

# Import the datetime module from the datetime library.
from datetime import datetime

# Print the beginning of the logging
print('Beginning Data Retreival')
print("-----------------------------")

#create counters

record_count = 1
set_count = 1

for i, city in enumerate(cities):

    #Group cities in sets of 50 for logging purposes.
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
        
        #print(f'Set count{set_count}') see how changes the set_count
    
    # Create endpoint URL with each city. city. 
    #replace is used to replace the s pace blanks to the cities that have 2 words
    #city_url = url + '&APPID=' + weather_api_key + '&q=' + city.replace(' ','+')
    city_url=f'{url}&APPID={weather_api_key}&q={city.replace(" ","+")}'
    #print(f'Set count{set_count}') see how the set works
    #print(f'Record count{record_count}') see how the record works

    # Log the URL, record, and set numbers and the city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    # Add 1 to the record count.
    record_count += 1
    
    try:
        # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()
        # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        weather_description=city_weather['weather'][0]['description']
        # Convert the date to ISO standard.
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        # Append the city information into city_data list.
        city_data.append({"City": city.title(),
                          "Country": city_country,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          'Current Description':weather_description})

# If an error is experienced, skip the city.
# the skipping cites are due to that we are generating random nums (lats and lngs)
# and may exists in citypy, since we are passing them to citypy
# but may not exists in Open Weather Map API.
    except:
        print("City not found. Skipping...", city)
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")

print(len(city_data))
```

```python
city_data_df=pd.DataFrame(city_data)
city_data_df
```

```python
output_data_file='WeatherPy_Database.csv'
city_data_df.to_csv(output_data_file, index_label='City_ID')
```
